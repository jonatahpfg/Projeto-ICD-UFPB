```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("rock_spotify.csv", error_bad_lines=False)
df_music = pd.read_csv("music.csv")
```

```python
df=df.dropna()
df_music= df_music.dropna()



medida1 = df.describe().drop(['count'], axis=0).drop(['index'], axis=1)
medida2 = df_music.describe().drop('count')

display(medida1)
display(medida2)
```

```python
df_numeric = df.select_dtypes(include='number')
for coluna in df_numeric.columns:
    plt.boxplot(df_numeric[coluna])
    plt.title(coluna)
    plt.show()
```

```python
df_music_numeric = df_music.select_dtypes(include='number')
for coluna in df_music_numeric.columns:
    plt.boxplot(df_music_numeric[coluna])
    plt.title(coluna)
    plt.show()
```
